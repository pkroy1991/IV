```python
import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import datetime as dt
from scipy.stats import norm
from scipy.interpolate import UnivariateSpline
from scipy.interpolate import griddata
from nsepython import nsefetch
from scipy import optimize
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
url = "https://www.nseindia.com/api/option-chain-indices?symbol=NIFTY"
data = nsefetch(url)
```

```python
call_data = []
for row in data['records']['data']:
    ce = row.get('CE')
    if ce:
        Strike = row['CE']['strikePrice']
        Expiry = row['CE']['expiryDate']
        Call_price = row['CE']['lastPrice']
        Spot_price = data['records']['underlyingValue']
        call_data.append({"Strike" : Strike,
                         "Expiry" : Expiry,
                         "Call_price" : Call_price,
                         "Spot_price" : Spot_price})
        
call_df = pd.DataFrame(call_data)
call_df = call_df.dropna(subset=['Strike', 'Spot_price', 'Call_price'])
call_df = call_df[call_df['Call_price']>0].copy()


def implied_vol_cal(V, S_0, X, r, T):
    if T<=0:
        return np.nan # option already expired

    intrinsik_value = max(S_0 - np.exp(-r*T)*X, 0)
    if V < intrinsik_value:
        return np.nan           # arbitrage
        
    def black_schole_price(x):
        d_1 = (np.log(S_0/X) + (r + ((x**2)/2))*T)/(x*np.sqrt(T))
        d_2 = d_1 - (x*np.sqrt(T))
    
        return (S_0*(norm.cdf(d_1)) - X*(np.exp((-r)*T)*norm.cdf(d_2)) - V)
        
    try:
        return optimize.brentq(black_schole_price, 1e-6, 5)
    except ValueError:
        return np.nan
       

iv_list = []
time_to_maturity_series = []
today = pd.Timestamp.today()

r = 0.065

for Strike, Spot_price, Call_price, Expiry in zip(call_df['Strike'], call_df['Spot_price'], call_df['Call_price'], call_df['Expiry']):
    expiry = pd.to_datetime(Expiry)
    time_to_maturity = expiry - today
    T = time_to_maturity/pd.Timedelta(days=365)
    time_to_maturity_series.append(T)
    iv = implied_vol_cal(Call_price, Spot_price, Strike, r, T)
    iv_list.append(iv)

call_df['IV'] = iv_list
call_df['Time_to_maturity'] = time_to_maturity_series

call_df = call_df[(call_df['IV'] > 0.01) & (call_df['IV']<3)]
```

# Term Structure feature for a given strike

```python
call_df.dropna(inplace = True)
spot = call_df['Spot_price'].iloc[0]
closest_strike = call_df.loc[(call_df['Strike']-spot).abs().idxmin(), 'Strike']

# strikes_to_plot = [call_df['Strike'].min(), closest_strike, call_df['Strike'].max()]   # Instead, consider closest_strike +- 10% strike range

strikes_to_plot = call_df[(call_df['Strike'] > closest_strike*(1-0.01)) & 
(call_df['Strike'] < closest_strike*(1+0.03))]['Strike'].unique().tolist()

for strike in strikes_to_plot:
    closest_group = call_df[call_df['Strike'] == strike]

    plt.plot(closest_group['Time_to_maturity'], closest_group['IV'], marker = 'o', label = f"Term structure feature for Strike {strike}")
    plt.xlabel('Time to maturity')
    plt.ylabel('Implied Volatility')
    plt.legend()
    plt.show()
```

# Log monenyness vs implied volatility

```python
call_df['log_moneyness'] = np.log(np.exp(r*call_df['Time_to_maturity'])*call_df['Spot_price']/call_df['Strike'])
# call_df = call_df[(call_df['Strike'] > 0.8*call_df['Spot_price']) & (call_df['Strike'] < 1.2*call_df['Spot_price'])]
call_df = call_df[(call_df['log_moneyness'] > np.log((1.2)**(-1))) & (call_df['log_moneyness'] < np.log((0.8)**(-1)))]

for expiry, group in call_df.groupby('Expiry'):
    group = group.sort_values('log_moneyness')
    x = group['log_moneyness']
    y = group['IV']
    if len(x) > 3:
        spline = UnivariateSpline(x, y, k=3, s=0.01)
    elif len(x) > 1:
        spline = UnivariateSpline(x, y, k=len(x)-1, s=0.01)
    else:
        continue
    polyline = np.linspace(x.iloc[0], x.iloc[-1], 200)
    plt.plot(polyline, spline(polyline), label='Cubic spline')
    plt.plot(x, y, marker = 'o', label = str(expiry))
    plt.xlabel('log moneyness log(F/X)')
    plt.ylabel('Implied Volatilty')
    plt.title('Implied volatilty smile by expiry')
    plt.legend()
    plt.grid(True)
    plt.show()
```

# Volatility surface

```python
x = call_df['log_moneyness']
y = call_df['Time_to_maturity']
z = call_df['IV']

xi = np.linspace(x.min(), x.max(), 50)
yi = np.linspace(y.min(), y.max(), 50)

X, Y = np.meshgrid(xi, yi)

Z = griddata((x, y), z, (X, Y), method = 'cubic')

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.grid()

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
ax.set_xlabel('log moneyness', labelpad=20)
ax.set_ylabel('Time to maturity', labelpad=20)
ax.set_zlabel('Implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=8)
plt.show()
```
